```python
url = "https://sysblok.ru/interviews/zlachnye-oblasti-nauki-mihail-gelfand-o-plagiate-lishenii-stepeni-i-nejeffektivnyh-chinovnikah/"

# Отправляем GET-запрос к указанному URL-адресу
response = requests.get(url)

# Создаем объект BeautifulSoup для парсинга HTML-кода страницы
soup = BeautifulSoup(response.content, 'html.parser')

# Находим элементы, содержащие информацию об авторах статьи
author_elements = soup.find_all("span", class_="author vcard")

# Создаем список для хранения имен авторов
authors = []

# Добавляем имена авторов в список
for author_element in author_elements:
    # Имя автора находится внутри тега <a>
    author_name = author_element.find("a").text.strip()
    authors.append(author_name)

# Выводим имена авторов
for author in authors:
    print(author)
```

```python
# Функция для получения текста статьи по ссылке
def get_article_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    # Извлекаем все абзацы из статьи
    paragraphs = soup.find_all('p')
    article_text = '\n'.join([p.get_text() for p in paragraphs])
    return article_text

# Получаем главную страницу сайта
url = 'https://sysblok.ru'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Извлекаем ссылки на статьи
links = soup.find_all('a')[:20]

# Обходим каждую ссылку
for link in links:
    article_url = link['href']
    # Если ссылка не на внешний ресурс, а на статью сайта sysblok.ru
    if article_url.startswith('/') and '/wp-content/' not in article_url:
        article_url = url + article_url
        article_text = get_article_text(article_url)
        print(article_text)
        print('---')
```

```python
# собираем все новости!
# полезно использовать try / except
        
for url in urls:
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        
        # Пример кода для извлечения заголовков статей
        articles = soup.find_all('h2', class_='entry-title')
        for article in articles:
            title = article.text.strip()
            print(title)
            
        # Пример кода для извлечения текста статей
        entries = soup.find_all('div', class_='entry-content')
        for entry in entries:
            text = entry.text.strip()
            print(text)
            
    except Exception as e:
        print(f"Ошибка при парсинге страницы {url}: {str(e)}")
```
